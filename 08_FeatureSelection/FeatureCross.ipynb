```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create synthetic data
np.random.seed(0)
class_one_q1 = np.random.randn(50, 2) + [3, 3]
class_two_q2 = np.random.randn(50, 2) + [-3, 3]
class_one_q3 = np.random.randn(50, 2) + [-3, -3]
class_two_q4 = np.random.randn(50, 2) + [3, -3]

# Combine the data
data = np.vstack([class_one_q1, class_two_q2, class_one_q3, class_two_q4])
labels = np.array([0]*50 + [1]*50 + [0]*50 + [1]*50)

# Create a dataframe
df = pd.DataFrame(data, columns=['X', 'Y'])
df['Class'] = labels

plt.figure(figsize=(10, 6))
plt.scatter(df["X"], df["Y"], c=df["Class"], cmap='coolwarm', s=100)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Synthetic Data - Feature cross')
plt.grid(True)
plt.show()
```

```python
def cl_fun(reg):
    if (reg.X >= 0) and (reg.Y >= 0):
        return "AC"
    elif (reg.X<0) and (reg.Y>=0):
        return "BC"
    elif (reg.X>=0) and (reg.Y<0):
        return "AD"
    else:
        return "BD"
```

```python
df["Category"] = df.apply(cl_fun, axis=1)   # Apply the function to the dataframe   
```
