```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
```

```python
df = pd.read_csv('../datasets/air-passengers.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df['Month'], df['Passengers'])
ax.set_xlabel('Date')
ax.set_ylabel('Number of air passengers')
plt.xticks(np.arange(0, 145, 12), np.arange(1949, 1962, 1))
fig.autofmt_xdate()
plt.tight_layout()

plt.show()
```

ACF (Sin diferenciar)

```python
ADF_result = adfuller(df['Passengers'])
print(f'ADF Statistic: {ADF_result[0]}')
print(f'p-value: {ADF_result[1]}')
```

La serie no es estacionaria pero igualmente grafiquemos el ACF...

```python
fig, (ax_acf, ax_pacf) = plt.subplots(2, 1, figsize=(10, 8))
ACF_plot = plot_acf(df['Passengers'], lags=40, ax=ax_acf)
PACF_plot = plot_pacf(df['Passengers'], lags=40, ax=ax_pacf)

plt.tight_layout()
plt.show()
```
