```python
#https://github.com/jamesqo/gun-violence-data/blob/master/DATA_01-2013_03-2018.tar.gz
import pandas as pd
import io
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
```

```python
idf = pd.read_csv("../../datasets/DATA_01-2013_03-2018.tar.gz", compression='gzip')
idf = idf[idf.columns.drop(["stage3.csv", 
                            "incident_url", 
                            "source_url", 
                            "incident_url_fields_missing", 
                            "sources", "notes", "state_house_district", "state_senate_district"])]
idf.replace({math.nan: None}, inplace=True)
```

```python

idf["date"] = pd.to_datetime(idf['date'])
```

```python
idf["latitude"] = idf["latitude"].astype(float)
idf["longitude"] = idf["longitude"].astype(float)
```

```python
#idf = idf.astype({"n_killed":"int", "n_injured":"int", 'n_guns_involved':'int'})
#Borro donde no hay lat y lon
idf = idf[(idf["latitude"].notnull()) & (idf["longitude"].notnull())]
```
