```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import OrdinalEncoder
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report
```

```python
# Definir los datos
data = {
    'Home Owner': ['Yes', 'No', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'No'],
    'Marital Status': ['Single', 'Married', 'Single', 'Married', 'Divorced', 'Married', 'Divorced', 'Single','Married', 'Single'],
    'Annual Income': [125,100,70,120,95,60,220,85,75,90],
    'Defaulted Borrower': ["No","No","No","No","Yes","No","No","Yes","No","Yes"]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Mostrar el DataFrame
print(df)
X = pd.get_dummies(df[df.columns.drop('Defaulted Borrower')])
y = df[['Defaulted Borrower']]
clf = DecisionTreeClassifier(max_depth=1, splitter='random', random_state=3)
clf.fit(X, y)
plt.figure(figsize=(10, 10))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True)
plt.show()
```
