```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import OrdinalEncoder
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report
# Crear el DataFrame con los datos
data = [
    ['Soleado', 'Muy Fuerte', "No"],
    ['Soleado', 'Fuerte', "Si" ],
    ['Soleado', 'Normal', "Si" ],
    ['Soleado', 'Nulo', "No" ],
    
    ['Nublado', 'Muy Fuerte', "No"],
    ['Nublado', 'Fuerte', "No" ],
    ['Nublado', 'Normal', "Si" ],
    ['Nublado', 'Nulo', "No" ],
    
    ['Lluvia', 'Muy Fuerte', "No"],
    ['Lluvia', 'Fuerte', "No" ],
    ['Lluvia', 'Normal', "Si" ],
    ['Lluvia', 'Nulo', "No" ],

]


df = pd.DataFrame(data, columns=['Cielo', 'Viento', 'Navega'])
df
```

```python
oe = OrdinalEncoder(categories=[['Soleado', 'Nublado', 'Lluvia'], ['Nulo', 'Normal',  'Fuerte', 'Muy Fuerte'], ['No', 'Si']], dtype=int)
df.loc[:, ["OCielo", "OViento", "ONavega"]] = oe.fit_transform(df)
```

```python
# Separar las caractersticas y las etiquetas
X = df[["OCielo", "OViento"]]
y = df['ONavega']
model = DecisionTreeClassifier(max_depth=2, criterion='gini')
model.fit(X, y)
# Crear y entrenar el rbol de decisin
# Visualizar el rbol de decisin
plt.figure(figsize=(10, 10))
plot_tree(model, feature_names=X.columns, class_names=oe.inverse_transform(df[["OCielo", "OViento", "ONavega"]])[:, 2], filled=True)
plt.show()
```

```python
y_pred = model.predict(X)
print(classification_report(y, y_pred))
```
